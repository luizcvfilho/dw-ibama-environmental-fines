# ETL para Data Warehouse - Multas Ambientais

## Modelo Dimensional (Star Schema)

### Nomenclatura de Chaves:
- **pk_*** : Primary Key (Chave Primária) nas dimensões
- **fk_*** : Foreign Key (Chave Estrangeira) na tabela fato

### Dimensões:
- **DTempo** (`pk_tempo`): Data da infração (ano, mês, trimestre, dia)
- **DInfrator** (`pk_infrator`): Informações sobre o infrator (documento, nome, tipo)
- **DLocal** (`pk_local`): Localização (UF, nome estado, município, região)
- **DInfracao** (`pk_infracao`): Tipo de infração e enquadramento legal
- **DDebito** (`pk_debito`): Informações sobre o débito (situação, moeda)

### Fato:
- **FAutoInfracao** (`pk_fato`): Tabela fato com as métricas e chaves estrangeiras
  - Chaves estrangeiras: `fk_tempo`, `fk_infrator`, `fk_local`, `fk_infracao`, `fk_debito`
  - Métricas: `valorOriginal`, `moedaOriginal`, `valorPadrao` (normalizado em Real)

````python
import os

estados = ['AC', 'AL', 'AM', 'AP', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 'MG', 'MS', 'MT', 
        'PA', 'PB', 'PE', 'PI', 'PR', 'RJ', 'RN', 'RO', 'RR', 'RS', 'SC', 'SE', 'SP', 'TO']

dfs = []
for estado in estados:
    arquivo = f"Dados/multasDistribuidasBensTutelados{estado}.csv"
    if os.path.exists(arquivo):
        df = pd.read_csv(arquivo, sep=";", encoding='utf-8')
        dfs.append(df)
        print(f"Carregado: {estado} - {len(df)} registros")

df_completo = pd.concat(dfs, ignore_index=True)
print(f"\nTotal de registros: {len(df_completo)}")
print(f"\nColunas: {df_completo.columns.tolist()}")
````

## 1. Criar Dimensão DTempo

````python
# Converter Data Auto para datetime
df_completo['Data Auto'] = pd.to_datetime(df_completo['Data Auto'], errors='coerce', dayfirst=True)

# Criar dimensão DTempo
DTempo = df_completo[['Data Auto']].drop_duplicates().dropna().copy()
DTempo['pk_tempo'] = range(1, len(DTempo) + 1)
DTempo['data'] = DTempo['Data Auto']
DTempo['ano'] = DTempo['Data Auto'].dt.year
DTempo['mes'] = DTempo['Data Auto'].dt.month
DTempo['dia'] = DTempo['Data Auto'].dt.day
DTempo['trimestre'] = DTempo['Data Auto'].dt.quarter
DTempo['diaSemana'] = DTempo['Data Auto'].dt.day_name()

# Selecionar colunas finais
DTempo = DTempo[['pk_tempo', 'data', 'ano', 'mes', 'dia', 'trimestre', 'diaSemana']]
DTempo = DTempo.sort_values('data').reset_index(drop=True)

print(f"Dimensão DTempo criada com {len(DTempo)} registros")
print("\nPrimeiras linhas:")
print(DTempo.head())
````

## 2. Criar Dimensão DInfrator

````python
# Criar dimensão DInfrator
DInfrator = df_completo[['CPF/CNPJ', 'Nome ou Razão Social']].drop_duplicates().copy()
DInfrator['pk_infrator'] = range(1, len(DInfrator) + 1)

# Determinar tipo de pessoa (CPF tem 11 dígitos, CNPJ tem 14)
DInfrator['documento'] = DInfrator['CPF/CNPJ'].astype(str).str.replace('.', '').str.replace('-', '').str.replace('/', '')
DInfrator['tipoPessoa'] = DInfrator['documento'].apply(lambda x: 'Pessoa Física' if len(x) == 11 else ('Pessoa Jurídica' if len(x) == 14 else 'Indefinido'))
DInfrator['nome'] = DInfrator['Nome ou Razão Social']

# Selecionar colunas finais
DInfrator = DInfrator[['pk_infrator', 'documento', 'nome', 'tipoPessoa']].reset_index(drop=True)

print(f"Dimensão DInfrator criada com {len(DInfrator)} registros")
print("\nDistribuição por tipo de pessoa:")
print(DInfrator['tipoPessoa'].value_counts())
print("\nPrimeiras linhas:")
print(DInfrator.head())
````

## 3. Criar Dimensão DLocal

````python
# Dicionário de regiões do Brasil
regioes = {
    'AC': 'Norte', 'AM': 'Norte', 'AP': 'Norte', 'PA': 'Norte', 'RO': 'Norte', 'RR': 'Norte', 'TO': 'Norte',
    'AL': 'Nordeste', 'BA': 'Nordeste', 'CE': 'Nordeste', 'MA': 'Nordeste', 'PB': 'Nordeste', 
    'PE': 'Nordeste', 'PI': 'Nordeste', 'RN': 'Nordeste', 'SE': 'Nordeste',
    'DF': 'Centro-Oeste', 'GO': 'Centro-Oeste', 'MT': 'Centro-Oeste', 'MS': 'Centro-Oeste',
    'ES': 'Sudeste', 'MG': 'Sudeste', 'RJ': 'Sudeste', 'SP': 'Sudeste',
    'PR': 'Sul', 'RS': 'Sul', 'SC': 'Sul'
}

# Dicionário de nomes dos estados
nomes_estados = {
    'AC': 'Acre', 'AL': 'Alagoas', 'AM': 'Amazonas', 'AP': 'Amapá',
    'BA': 'Bahia', 'CE': 'Ceará', 'DF': 'Distrito Federal', 'ES': 'Espírito Santo',
    'GO': 'Goiás', 'MA': 'Maranhão', 'MG': 'Minas Gerais', 'MS': 'Mato Grosso do Sul',
    'MT': 'Mato Grosso', 'PA': 'Pará', 'PB': 'Paraíba', 'PE': 'Pernambuco',
    'PI': 'Piauí', 'PR': 'Paraná', 'RJ': 'Rio de Janeiro', 'RN': 'Rio Grande do Norte',
    'RO': 'Rondônia', 'RR': 'Roraima', 'RS': 'Rio Grande do Sul', 'SC': 'Santa Catarina',
    'SE': 'Sergipe', 'SP': 'São Paulo', 'TO': 'Tocantins'
}

# Criar dimensão DLocal
DLocal = df_completo[['UF', 'Município']].drop_duplicates().copy()
DLocal['pk_local'] = range(1, len(DLocal) + 1)
DLocal['uf'] = DLocal['UF']
DLocal['nome_estado'] = DLocal['UF'].map(nomes_estados)
DLocal['municipio'] = DLocal['Município']
DLocal['regiao'] = DLocal['UF'].map(regioes)

# Selecionar colunas finais
DLocal = DLocal[['pk_local', 'uf', 'nome_estado', 'municipio', 'regiao']].reset_index(drop=True)

print(f"Dimensão DLocal criada com {len(DLocal)} registros")
print("\nDistribuição por região:")
print(DLocal['regiao'].value_counts())
print("\nPrimeiras linhas:")
print(DLocal.head())
````

## 4. Criar Dimensão DInfracao

````python
# Criar dimensão DInfracao
DInfracao = df_completo[['Tipo Auto', 'Tipo Infração', 'Enquadramento Legal']].drop_duplicates().copy()
DInfracao['pk_infracao'] = range(1, len(DInfracao) + 1)
DInfracao['tipoAuto'] = DInfracao['Tipo Auto']
DInfracao['tipoInfracao'] = DInfracao['Tipo Infração']
DInfracao['enquadramentoLegal'] = DInfracao['Enquadramento Legal']

# Selecionar colunas finais
DInfracao = DInfracao[['pk_infracao', 'tipoAuto', 'tipoInfracao', 'enquadramentoLegal']].reset_index(drop=True)

print(f"Dimensão DInfracao criada com {len(DInfracao)} registros")
print("\nTipos de Auto:")
print(df_completo['Tipo Auto'].value_counts())
print("\nPrimeiras linhas:")
print(DInfracao.head(10))
````

## 5. Criar Dimensão DDebito

````python
# Criar dimensão DDebito
DDebito = df_completo[['Situação Débito', 'Moeda']].drop_duplicates().copy()
DDebito['pk_debito'] = range(1, len(DDebito) + 1)
DDebito['situacao'] = DDebito['Situação Débito']
DDebito['moeda'] = DDebito['Moeda']

# Selecionar colunas finais
DDebito = DDebito[['pk_debito', 'situacao', 'moeda']].reset_index(drop=True)

print(f"Dimensão DDebito criada com {len(DDebito)} registros")
print("\nSituações de débito:")
print(df_completo['Situação Débito'].value_counts())
print("\nPrimeiras linhas:")
print(DDebito.head())
````

## 6. Criar Tabela Fato FAutoInfracao

````python
# Preparar dataframe base para fato
FAutoInfracao = df_completo.copy()

# Fazer merge com DTempo
FAutoInfracao = FAutoInfracao.merge(
    DTempo[['data', 'pk_tempo']], 
    left_on='Data Auto', 
    right_on='data', 
    how='left'
).drop('data', axis=1)

# Fazer merge com DInfrator - corrigido
# Preparar colunas temporárias para o merge
FAutoInfracao['cpf_cnpj_temp'] = FAutoInfracao['CPF/CNPJ'].astype(str).str.replace('.', '').str.replace('-', '').str.replace('/', '')
FAutoInfracao['nome_temp'] = FAutoInfracao['Nome ou Razão Social']

FAutoInfracao = FAutoInfracao.merge(
    DInfrator[['documento', 'nome', 'pk_infrator']], 
    left_on=['cpf_cnpj_temp', 'nome_temp'],
    right_on=['documento', 'nome'],
    how='left'
).drop(['cpf_cnpj_temp', 'nome_temp', 'documento', 'nome'], axis=1)

# Fazer merge com DLocal
FAutoInfracao = FAutoInfracao.merge(
    DLocal[['uf', 'municipio', 'pk_local']], 
    left_on=['UF', 'Município'], 
    right_on=['uf', 'municipio'], 
    how='left'
).drop(['uf', 'municipio'], axis=1)

# Fazer merge com DInfracao
FAutoInfracao = FAutoInfracao.merge(
    DInfracao[['tipoAuto', 'tipoInfracao', 'enquadramentoLegal', 'pk_infracao']], 
    left_on=['Tipo Auto', 'Tipo Infração', 'Enquadramento Legal'], 
    right_on=['tipoAuto', 'tipoInfracao', 'enquadramentoLegal'], 
    how='left'
).drop(['tipoAuto', 'tipoInfracao', 'enquadramentoLegal'], axis=1)

# Fazer merge com DDebito
FAutoInfracao = FAutoInfracao.merge(
    DDebito[['situacao', 'moeda', 'pk_debito']], 
    left_on=['Situação Débito', 'Moeda'], 
    right_on=['situacao', 'moeda'], 
    how='left'
).drop(['situacao', 'moeda'], axis=1)

print("Merge completo! Verificando chaves estrangeiras...")
print(f"Registros sem fk_tempo: {FAutoInfracao['pk_tempo'].isna().sum()}")
print(f"Registros sem fk_infrator: {FAutoInfracao['pk_infrator'].isna().sum()}")
print(f"Registros sem fk_local: {FAutoInfracao['pk_local'].isna().sum()}")
print(f"Registros sem fk_infracao: {FAutoInfracao['pk_infracao'].isna().sum()}")
print(f"Registros sem fk_debito: {FAutoInfracao['pk_debito'].isna().sum()}")
````

````python
# Investigar os registros sem fk_tempo
print("="*80)
print("INVESTIGAÇÃO: Registros sem fk_tempo")
print("="*80)

registros_sem_tempo = FAutoInfracao[FAutoInfracao['pk_tempo'].isna()]

print(f"\nTotal de registros sem fk_tempo: {len(registros_sem_tempo)}")
print("\nColunas relevantes desses registros:")
print(registros_sem_tempo[['Nº AI', 'Data Auto', 'Nome ou Razão Social', 'UF', 'Município', 'Tipo Auto']].to_string())

print("\n" + "-"*80)
print("Análise da coluna 'Data Auto':")
print(registros_sem_tempo['Data Auto'].to_string())

print("\n" + "-"*80)
print("Valores originais da data (antes da conversão):")
# Pegar os mesmos registros do df_completo original
indices_problema = registros_sem_tempo.index
dados_originais = df_completo.loc[indices_problema, ['Nº AI', 'Data Auto', 'UF', 'Município']]
print(dados_originais.to_string())

print("\n" + "="*80)
````

````python
# Verificar quais moedas existem nos dados
print("Moedas encontradas nos dados:")
print(df_completo['Moeda'].value_counts())
print("\n" + "="*80)
````

````python
taxas_conversao = {
    'Real': 1.0,
    'Cruzeiro Real': 1 / 2_750.0,
    'Cruzeiro (90 a 93)': 1 / 2_750_000.0,
    'Cruzado Novo': 1 / 2_750_000.0,
    'Cruzado': 1 / 2_750_000_000.0,
    'Cruzeiro (70 a 86)': 1 / 2_750_000_000_000.0,
    'UFIR': 1.0641,
    'MVR': 2_266.17 / 2_750_000.0,
    'BTN': 126.8621 / 2_750_000.0,
    'OTN': 126.8621 * 6.17 / 2_750_000.0
}

def converter_para_real(row):
    try:
        valor_str = str(row['Valor do Auto']).replace('.', '').replace('R$', '').strip()
        valor_str = valor_str.replace(',', '.')
        valor_num = pd.to_numeric(valor_str, errors='coerce')
        
        if pd.isna(valor_num):
            return None

        moeda = row['Moeda']
        taxa = taxas_conversao.get(moeda, 1.0)

        valor_real = valor_num * taxa
        
        return valor_real
    except Exception:
        return None

print("Taxas de conversão definidas:")
for moeda, taxa in taxas_conversao.items():
    print(f"  {moeda}: {taxa}")
print("\n" + "="*80)
````

````python
print("Convertendo valores para Real...")
FAutoInfracao['valorPadrao'] = FAutoInfracao.apply(converter_para_real, axis=1)
print(f"Conversão concluída! {FAutoInfracao['valorPadrao'].notna().sum()} valores convertidos com sucesso.")

FAutoInfracao_final = FAutoInfracao[[
    'Nº AI',
    'pk_tempo',
    'pk_infrator',
    'pk_local',
    'pk_infracao',
    'pk_debito',
    'Valor do Auto',
    'Moeda',
    'valorPadrao',
]].copy()

# Renomear colunas
FAutoInfracao_final.columns = [
    'numAuto', 'fk_tempo', 'fk_infrator', 'fk_local', 'fk_infracao', 
    'fk_debito', 'valorOriginal', 'moedaOriginal', 'valorPadrao'
]

# Adicionar surrogate key
FAutoInfracao_final['pk_fato'] = range(1, len(FAutoInfracao_final) + 1)

# Reordenar colunas
FAutoInfracao_final = FAutoInfracao_final[[
    'pk_fato', 'numAuto', 'fk_tempo', 'fk_infrator', 'fk_local', 
    'fk_infracao', 'fk_debito', 'valorOriginal', 'moedaOriginal', 'valorPadrao'
]]

print(f"Tabela Fato FAutoInfracao criada com {len(FAutoInfracao_final)} registros")
print("\nEstatísticas do valor das multas (convertido para Real):")
print(FAutoInfracao_final['valorPadrao'].describe())
print(f"\nValor total em Real: R$ {FAutoInfracao_final['valorPadrao'].sum():,.2f}")
print("\nDistribuição por moeda original:")
print(FAutoInfracao_final['moedaOriginal'].value_counts())
print("\nPrimeiras linhas:")
print(FAutoInfracao_final.head())
````

## 9. Exemplo de Consultas Analíticas

Agora você pode fazer análises complexas usando o modelo dimensional!

````python
# Exemplo 1: Total de multas por região
analise_regiao = FAutoInfracao_final.merge(DLocal, left_on='fk_local', right_on='pk_local').groupby('regiao').agg({
    'pk_fato': 'count',
    'valorPadrao': 'sum'
}).round(2)
analise_regiao.columns = ['Quantidade', 'Valor Total (R$)']
print("📊 Total de Multas por Região:")
print(analise_regiao.sort_values('Quantidade', ascending=False))

# Exemplo 2: Multas por ano
analise_ano = FAutoInfracao_final.merge(DTempo, left_on='fk_tempo', right_on='pk_tempo').groupby('ano').agg({
    'pk_fato': 'count',
    'valorPadrao': 'sum'
}).round(2)
analise_ano.columns = ['Quantidade', 'Valor Total (R$)']
print("\n📅 Total de Multas por Ano (últimos 10 anos):")
print(analise_ano.sort_index(ascending=False).head(10))

# Exemplo 3: Top 5 tipos de infração
analise_infracao = FAutoInfracao_final.merge(DInfracao, left_on='fk_infracao', right_on='pk_infracao').groupby('tipoInfracao').agg({
    'pk_fato': 'count',
    'valorPadrao': 'sum'
}).round(2)
analise_infracao.columns = ['Quantidade', 'Valor Total (R$)']
print("\n⚖️ Top 5 Tipos de Infração:")
print(analise_infracao.sort_values('Quantidade', ascending=False).head())
````

                    └────────┬─────────┘

       ┌─────────────────────┼─────────────────────┐
       │                     │                     │
┌──────┴────────┐     ┌──────┴──────────────┐  ┌──┴────────────┐
│  DInfrator    │     │  FAutoInfracao      │  │    DLocal     │
├───────────────┤     ├─────────────────────┤  ├───────────────┤
│pk_infrator PK │─────│ pk_fato PK          │──│ pk_local PK   │
│documento      │     │ numAuto             │  │ uf            │
│nome           │     │ fk_tempo FK         │  │ nome_estado   │
