# ETL para Data Warehouse - Multas Ambientais

## Modelo Dimensional (Star Schema)

### Nomenclatura de Chaves:
- **pk_*** : Primary Key (Chave Prim√°ria) nas dimens√µes
- **fk_*** : Foreign Key (Chave Estrangeira) na tabela fato

### Dimens√µes:
- **DTempo** (`pk_tempo`): Data da infra√ß√£o (ano, m√™s, trimestre, dia)
- **DInfrator** (`pk_infrator`): Informa√ß√µes sobre o infrator (documento, nome, tipo)
- **DLocal** (`pk_local`): Localiza√ß√£o (UF, nome estado, munic√≠pio, regi√£o)
- **DInfracao** (`pk_infracao`): Tipo de infra√ß√£o e enquadramento legal
- **DDebito** (`pk_debito`): Informa√ß√µes sobre o d√©bito (situa√ß√£o, moeda)

### Fato:
- **FAutoInfracao**:
  - Chaves estrangeiras: `fk_tempo`, `fk_infrator`, `fk_local`, `fk_infracao`, `fk_debito`
  - M√©tricas: `numAuto`, `valorOriginal`, `moedaOriginal`, `valorPadrao` (normalizado em Real)
  - A chave da tabela fato √© composta pela combina√ß√£o das FKs

````python
import os

estados = ['AC', 'AL', 'AM', 'AP', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 'MG', 'MS', 'MT', 
        'PA', 'PB', 'PE', 'PI', 'PR', 'RJ', 'RN', 'RO', 'RR', 'RS', 'SC', 'SE', 'SP', 'TO']

dfs = []
for estado in estados:
    arquivo = f"Dados/multasDistribuidasBensTutelados{estado}.csv"
    if os.path.exists(arquivo):
        df = pd.read_csv(arquivo, sep=";", encoding='utf-8')
        dfs.append(df)
        print(f"Carregado: {estado} - {len(df)} registros")

df_completo = pd.concat(dfs, ignore_index=True)
print(f"\nTotal de registros: {len(df_completo)}")
print(f"\nColunas: {df_completo.columns.tolist()}")
````

````python
print("="*80)
print("AN√ÅLISE DE VALORES INCONSISTENTES")
print("="*80)

# An√°lise de datas
if 'Data Auto' in df_completo.columns:
    print("\nüìÖ An√°lise de Datas:")
    
    # Verificar exemplos de datas
    for data in df_completo['Data Auto'].unique()[:10]:
        if pd.notna(data):
            print(f"  Exemplo de data: {data}")
            break
    
    # Verificar datas com formato suspeito
    print("\n  Verificando datas inconsistentes...")
    datas_com_erro = df_completo[~df_completo['Data Auto'].astype(str).str.contains(r'\d{4}', na=False)]
    if len(datas_com_erro) > 0:
        print(f"  ‚ö†Ô∏è {len(datas_com_erro)} datas com formato suspeito")

# An√°lise de valores monet√°rios
if 'Valor do Auto' in df_completo.columns:
    print("\nüí∞ An√°lise de Valores:")
    
    # Converter para num√©rico para an√°lise
    df_completo['Valor_Temp'] = df_completo['Valor do Auto'].astype(str).str.replace('.', '').str.replace('R$', '').str.strip()
    df_completo['Valor_Temp'] = df_completo['Valor_Temp'].str.replace(',', '.')
    df_completo['Valor_Temp'] = pd.to_numeric(df_completo['Valor_Temp'], errors='coerce')
    
    valores_nulos = df_completo['Valor_Temp'].isnull().sum()
    valores_zero = (df_completo['Valor_Temp'] == 0).sum()
    valores_negativos = (df_completo['Valor_Temp'] < 0).sum()
    
    print(f"  üìä Valores nulos ou inv√°lidos: {valores_nulos:,}")
    print(f"  üî¥ Valores iguais a zero: {valores_zero:,}")
    print(f"  ‚õî Valores negativos: {valores_negativos:,}")
    
    if valores_nulos + valores_zero + valores_negativos > 0:
        print(f"\n  ‚ö†Ô∏è Total de valores problem√°ticos: {valores_nulos + valores_zero + valores_negativos:,}")
    
    # Estat√≠sticas descritivas
    print("\n  üìà Estat√≠sticas dos valores:")
    print(f"     M√≠nimo: R$ {df_completo['Valor_Temp'].min():,.2f}")
    print(f"     M√°ximo: R$ {df_completo['Valor_Temp'].max():,.2f}")
    print(f"     M√©dia: R$ {df_completo['Valor_Temp'].mean():,.2f}")
    print(f"     Mediana: R$ {df_completo['Valor_Temp'].median():,.2f}")
    
    # Limpar coluna tempor√°ria
    df_completo.drop('Valor_Temp', axis=1, inplace=True)


print("\n" + "="*80)
````

````python
print("="*80)
print("RESUMO GERAL DA QUALIDADE DOS DADOS")
print("="*80)

total_registros = len(df_completo)
total_colunas = len(df_completo.columns)
total_celulas = total_registros * total_colunas
celulas_nulas = df_completo.isnull().sum().sum()
celulas_preenchidas = total_celulas - celulas_nulas

print("\nüìà M√©tricas Gerais:")
print(f"  ‚Ä¢ Total de Registros: {total_registros:,}")
print(f"  ‚Ä¢ Total de Colunas: {total_colunas}")
print(f"  ‚Ä¢ Total de C√©lulas: {total_celulas:,}")
print(f"  ‚Ä¢ C√©lulas Preenchidas: {celulas_preenchidas:,}")
print(f"  ‚Ä¢ C√©lulas Nulas: {celulas_nulas:,}")

completude_geral = (celulas_preenchidas / total_celulas) * 100
print(f"\n‚ú® Completude Geral: {completude_geral:.2f}%")

# Classifica√ß√£o da Qualidade
if completude_geral >= 95:
    qualidade = "‚úÖ Excelente"
elif completude_geral >= 85:
    qualidade = "‚úì Boa"
elif completude_geral >= 70:
    qualidade = "‚ö†Ô∏è Regular"
else:
    qualidade = "‚ùå Baixa"

print(f"üìä Avalia√ß√£o de Qualidade: {qualidade}")
````

## 1. Criar Dimens√£o DTempo

````python
# Converter Data Auto para datetime
df_completo['Data Auto'] = pd.to_datetime(df_completo['Data Auto'], errors='coerce', dayfirst=True)

# Criar dimens√£o DTempo
DTempo = df_completo[['Data Auto']].drop_duplicates().dropna().copy()
DTempo['pk_tempo'] = range(1, len(DTempo) + 1)
DTempo['data'] = DTempo['Data Auto']
DTempo['ano'] = DTempo['Data Auto'].dt.year
DTempo['mes'] = DTempo['Data Auto'].dt.month
DTempo['dia'] = DTempo['Data Auto'].dt.day
DTempo['trimestre'] = DTempo['Data Auto'].dt.quarter
DTempo['diaSemana'] = DTempo['Data Auto'].dt.day_name()

# Selecionar colunas finais
DTempo = DTempo[['pk_tempo', 'data', 'ano', 'mes', 'dia', 'trimestre', 'diaSemana']]
DTempo = DTempo.sort_values('data').reset_index(drop=True)

print(f"Dimens√£o DTempo criada com {len(DTempo)} registros")
print("\nPrimeiras linhas:")
print(DTempo.head())
````

## 2. Criar Dimens√£o DInfrator

````python
# Criar dimens√£o DInfrator
DInfrator = df_completo[['CPF/CNPJ', 'Nome ou Raz√£o Social']].drop_duplicates().copy()
DInfrator['pk_infrator'] = range(1, len(DInfrator) + 1)

# Determinar tipo de pessoa (CPF tem 11 d√≠gitos, CNPJ tem 14)
DInfrator['documento'] = DInfrator['CPF/CNPJ'].astype(str).str.replace('.', '').str.replace('-', '').str.replace('/', '')
DInfrator['tipoPessoa'] = DInfrator['documento'].apply(lambda x: 'Pessoa F√≠sica' if len(x) == 11 else ('Pessoa Jur√≠dica' if len(x) == 14 else 'Indefinido'))
DInfrator['nome'] = DInfrator['Nome ou Raz√£o Social']

# Selecionar colunas finais
DInfrator = DInfrator[['pk_infrator', 'documento', 'nome', 'tipoPessoa']].reset_index(drop=True)

print(f"Dimens√£o DInfrator criada com {len(DInfrator)} registros")
print("\nDistribui√ß√£o por tipo de pessoa:")
print(DInfrator['tipoPessoa'].value_counts())
print("\nPrimeiras linhas:")
print(DInfrator.head())
````

## 3. Criar Dimens√£o DLocal

````python
# Dicion√°rio de regi√µes do Brasil
regioes = {
    'AC': 'Norte', 'AM': 'Norte', 'AP': 'Norte', 'PA': 'Norte', 'RO': 'Norte', 'RR': 'Norte', 'TO': 'Norte',
    'AL': 'Nordeste', 'BA': 'Nordeste', 'CE': 'Nordeste', 'MA': 'Nordeste', 'PB': 'Nordeste', 
    'PE': 'Nordeste', 'PI': 'Nordeste', 'RN': 'Nordeste', 'SE': 'Nordeste',
    'DF': 'Centro-Oeste', 'GO': 'Centro-Oeste', 'MT': 'Centro-Oeste', 'MS': 'Centro-Oeste',
    'ES': 'Sudeste', 'MG': 'Sudeste', 'RJ': 'Sudeste', 'SP': 'Sudeste',
    'PR': 'Sul', 'RS': 'Sul', 'SC': 'Sul'
}

# Dicion√°rio de nomes dos estados
nomes_estados = {
    'AC': 'Acre', 'AL': 'Alagoas', 'AM': 'Amazonas', 'AP': 'Amap√°',
    'BA': 'Bahia', 'CE': 'Cear√°', 'DF': 'Distrito Federal', 'ES': 'Esp√≠rito Santo',
    'GO': 'Goi√°s', 'MA': 'Maranh√£o', 'MG': 'Minas Gerais', 'MS': 'Mato Grosso do Sul',
    'MT': 'Mato Grosso', 'PA': 'Par√°', 'PB': 'Para√≠ba', 'PE': 'Pernambuco',
    'PI': 'Piau√≠', 'PR': 'Paran√°', 'RJ': 'Rio de Janeiro', 'RN': 'Rio Grande do Norte',
    'RO': 'Rond√¥nia', 'RR': 'Roraima', 'RS': 'Rio Grande do Sul', 'SC': 'Santa Catarina',
    'SE': 'Sergipe', 'SP': 'S√£o Paulo', 'TO': 'Tocantins'
}

# Criar dimens√£o DLocal
DLocal = df_completo[['UF', 'Munic√≠pio']].drop_duplicates().copy()
DLocal['pk_local'] = range(1, len(DLocal) + 1)
DLocal['uf'] = DLocal['UF']
DLocal['nome_estado'] = DLocal['UF'].map(nomes_estados)
DLocal['municipio'] = DLocal['Munic√≠pio']
DLocal['regiao'] = DLocal['UF'].map(regioes)

# Selecionar colunas finais
DLocal = DLocal[['pk_local', 'uf', 'nome_estado', 'municipio', 'regiao']].reset_index(drop=True)

print(f"Dimens√£o DLocal criada com {len(DLocal)} registros")
print("\nDistribui√ß√£o por regi√£o:")
print(DLocal['regiao'].value_counts())
print("\nPrimeiras linhas:")
print(DLocal.head())
````

## 4. Criar Dimens√£o DInfracao

````python
# Criar dimens√£o DInfracao
DInfracao = df_completo[['Tipo Auto', 'Tipo Infra√ß√£o', 'Enquadramento Legal']].drop_duplicates().copy()
DInfracao['pk_infracao'] = range(1, len(DInfracao) + 1)
DInfracao['tipoAuto'] = DInfracao['Tipo Auto']
DInfracao['tipoInfracao'] = DInfracao['Tipo Infra√ß√£o']
DInfracao['enquadramentoLegal'] = DInfracao['Enquadramento Legal']

# Selecionar colunas finais
DInfracao = DInfracao[['pk_infracao', 'tipoAuto', 'tipoInfracao', 'enquadramentoLegal']].reset_index(drop=True)

print(f"Dimens√£o DInfracao criada com {len(DInfracao)} registros")
print("\nTipos de Auto:")
print(df_completo['Tipo Auto'].value_counts())
print("\nPrimeiras linhas:")
print(DInfracao.head(10))
````

## 5. Criar Dimens√£o DDebito

````python
# Criar dimens√£o DDebito
DDebito = df_completo[['Situa√ß√£o D√©bito', 'Moeda']].drop_duplicates().copy()
DDebito['pk_debito'] = range(1, len(DDebito) + 1)
DDebito['situacao'] = DDebito['Situa√ß√£o D√©bito']
DDebito['moeda'] = DDebito['Moeda']

# Selecionar colunas finais
DDebito = DDebito[['pk_debito', 'situacao', 'moeda']].reset_index(drop=True)

print(f"Dimens√£o DDebito criada com {len(DDebito)} registros")
print("\nSitua√ß√µes de d√©bito:")
print(df_completo['Situa√ß√£o D√©bito'].value_counts())
print("\nPrimeiras linhas:")
print(DDebito.head())
````

## 6. Criar Tabela Fato FAutoInfracao

````python
# Preparar dataframe base para fato
FAutoInfracao = df_completo.copy()

# Fazer merge com DTempo
FAutoInfracao = FAutoInfracao.merge(
    DTempo[['data', 'pk_tempo']], 
    left_on='Data Auto', 
    right_on='data', 
    how='left'
).drop('data', axis=1)

# Fazer merge com DInfrator - corrigido
# Preparar colunas tempor√°rias para o merge
FAutoInfracao['cpf_cnpj_temp'] = FAutoInfracao['CPF/CNPJ'].astype(str).str.replace('.', '').str.replace('-', '').str.replace('/', '')
FAutoInfracao['nome_temp'] = FAutoInfracao['Nome ou Raz√£o Social']

FAutoInfracao = FAutoInfracao.merge(
    DInfrator[['documento', 'nome', 'pk_infrator']], 
    left_on=['cpf_cnpj_temp', 'nome_temp'],
    right_on=['documento', 'nome'],
    how='left'
).drop(['cpf_cnpj_temp', 'nome_temp', 'documento', 'nome'], axis=1)

# Fazer merge com DLocal
FAutoInfracao = FAutoInfracao.merge(
    DLocal[['uf', 'municipio', 'pk_local']], 
    left_on=['UF', 'Munic√≠pio'], 
    right_on=['uf', 'municipio'], 
    how='left'
).drop(['uf', 'municipio'], axis=1)

# Fazer merge com DInfracao
FAutoInfracao = FAutoInfracao.merge(
    DInfracao[['tipoAuto', 'tipoInfracao', 'enquadramentoLegal', 'pk_infracao']], 
    left_on=['Tipo Auto', 'Tipo Infra√ß√£o', 'Enquadramento Legal'], 
    right_on=['tipoAuto', 'tipoInfracao', 'enquadramentoLegal'], 
    how='left'
).drop(['tipoAuto', 'tipoInfracao', 'enquadramentoLegal'], axis=1)

# Fazer merge com DDebito
FAutoInfracao = FAutoInfracao.merge(
    DDebito[['situacao', 'moeda', 'pk_debito']], 
    left_on=['Situa√ß√£o D√©bito', 'Moeda'], 
    right_on=['situacao', 'moeda'], 
    how='left'
).drop(['situacao', 'moeda'], axis=1)

print("Merge completo! Verificando chaves estrangeiras...")
print(f"Registros sem fk_tempo: {FAutoInfracao['pk_tempo'].isna().sum()}")
print(f"Registros sem fk_infrator: {FAutoInfracao['pk_infrator'].isna().sum()}")
print(f"Registros sem fk_local: {FAutoInfracao['pk_local'].isna().sum()}")
print(f"Registros sem fk_infracao: {FAutoInfracao['pk_infracao'].isna().sum()}")
print(f"Registros sem fk_debito: {FAutoInfracao['pk_debito'].isna().sum()}")
````

````python
# Verificar quais moedas existem nos dados
print("Moedas encontradas nos dados:")
print(df_completo['Moeda'].value_counts())
print("\n" + "="*80)
````

````python
taxas_conversao = {
    'Real': 1.0,
    'Cruzeiro Real': 1 / 2_750.0,
    'Cruzeiro (90 a 93)': 1 / 2_750_000.0,
    'Cruzado Novo': 1 / 2_750_000.0,
    'Cruzado': 1 / 2_750_000_000.0,
    'Cruzeiro (70 a 86)': 1 / 2_750_000_000_000.0,
    'UFIR': 1.0641,
    'MVR': 2_266.17 / 2_750_000.0,
    'BTN': 126.8621 / 2_750_000.0,
    'OTN': 126.8621 * 6.17 / 2_750_000.0
}

def converter_para_real(row):
    try:
        valor_str = str(row['Valor do Auto']).replace('.', '').replace('R$', '').strip()
        valor_str = valor_str.replace(',', '.')
        valor_num = pd.to_numeric(valor_str, errors='coerce')
        
        if pd.isna(valor_num):
            return None

        moeda = row['Moeda']
        taxa = taxas_conversao.get(moeda, 1.0)

        valor_real = valor_num * taxa
        
        return valor_real
    except Exception:
        return None

print("Taxas de convers√£o definidas:")
for moeda, taxa in taxas_conversao.items():
    print(f"  {moeda}: {taxa}")
print("\n" + "="*80)
````

````python
print("Convertendo valores para Real...")
FAutoInfracao['valorPadrao'] = FAutoInfracao.apply(converter_para_real, axis=1)
print(f"Convers√£o conclu√≠da! {FAutoInfracao['valorPadrao'].notna().sum()} valores convertidos com sucesso.")

FAutoInfracao_final = FAutoInfracao[[
    'N¬∫ AI',
    'pk_tempo',
    'pk_infrator',
    'pk_local',
    'pk_infracao',
    'pk_debito',
    'Valor do Auto',
    'Moeda',
    'valorPadrao',
]].copy()

# Renomear colunas
FAutoInfracao_final.columns = [
    'numAuto', 'fk_tempo', 'fk_infrator', 'fk_local', 'fk_infracao', 
    'fk_debito', 'valorOriginal', 'moedaOriginal', 'valorPadrao'
]

print("Tabela Fato FAutoInfracao criada com {len(FAutoInfracao_final)} registros")
print("\nüìã Estrutura da tabela fato (sem PK, apenas FKs + m√©tricas):")
print("   Chaves Estrangeiras: fk_tempo, fk_infrator, fk_local, fk_infracao, fk_debito")
print("   M√©tricas: numAuto, valorOriginal, moedaOriginal, valorPadrao")
print("\nEstat√≠sticas do valor das multas (convertido para Real):")
print(FAutoInfracao_final['valorPadrao'].describe())
print(f"\nValor total em Real: R$ {FAutoInfracao_final['valorPadrao'].sum():,.2f}")
print("\nDistribui√ß√£o por moeda original:")
print(FAutoInfracao_final['moedaOriginal'].value_counts())
print("\nPrimeiras linhas:")
print(FAutoInfracao_final.head())
````

````python
# Exemplo 1: Total de multas por regi√£o
analise_regiao = FAutoInfracao_final.merge(DLocal, left_on='fk_local', right_on='pk_local').groupby('regiao').agg({
    'numAuto': 'count',
    'valorPadrao': 'sum'
}).round(2)
analise_regiao.columns = ['Quantidade', 'Valor Total (R$)']
print("üìä Total de Multas por Regi√£o:")
print(analise_regiao.sort_values('Quantidade', ascending=False))

# Exemplo 2: Multas por ano
analise_ano = FAutoInfracao_final.merge(DTempo, left_on='fk_tempo', right_on='pk_tempo').groupby('ano').agg({
    'numAuto': 'count',
    'valorPadrao': 'sum'
}).round(2)
analise_ano.columns = ['Quantidade', 'Valor Total (R$)']
print("\nüìÖ Total de Multas por Ano (√∫ltimos 10 anos):")
print(analise_ano.sort_index(ascending=False).head(10))

# Exemplo 3: Top 5 tipos de infra√ß√£o
analise_infracao = FAutoInfracao_final.merge(DInfracao, left_on='fk_infracao', right_on='pk_infracao').groupby('tipoInfracao').agg({
    'numAuto': 'count',
    'valorPadrao': 'sum'
}).round(2)
analise_infracao.columns = ['Quantidade', 'Valor Total (R$)']
print("\n‚öñÔ∏è Top 5 Tipos de Infra√ß√£o:")
print(analise_infracao.sort_values('Quantidade', ascending=False).head())
````
